```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('regular.csv', encoding = 'latin-1')
```

```python
tripleros = df.sort_values(by = '3P', ascending = False).head(5)
tripleros = tripleros.sort_values(by = '3P', ascending = True)
triplerosx = tripleros['Player']
triplerosy = tripleros['3P']

fig1 = plt.figure(1)
fig1.set_figheight(10)
fig1.set_figwidth(15)
fig1.set_facecolor('white')
ax = fig1.add_subplot(1,1,1)
plt.xlim(3, 4.7)
plt.title('Jugadores con m√°s cantidad de triples por partido - Temporada regular 2021-2022')
plt.xlabel('Triples por partido')
plt.ylabel('Jugador')
ax.barh(triplerosx, triplerosy, color = '#4CAF50')
plt.savefig("Tripleros.png")
```

```python
#Posici√≥n con mejor conversi√≥n de triples

posicion_tripleros1 = df.groupby(['Pos']).mean().reset_index()
posicion_tripleros1 = posicion_tripleros1[['Pos', '3P%']].sort_values(by = '3P%', ascending = False)
posicion_tripleros1 = posicion_tripleros1.sort_values(by='3P%', ascending=True)


posicion = posicion_tripleros1['Pos']
conversiontriples = posicion_tripleros1['3P%']

fig2 = plt.figure(1)
fig2.set_figheight(10)
fig2.set_figwidth(8)
plt.xlim(0.15, 0.40)
fig2.set_facecolor('white')
plt.title('Posiciones con mayor conversi√≥n de triples - Temporada regular 2021-2022')
plt.xlabel('Conversi√≥n de triples')
plt.ylabel('Posici√≥n')
plt.barh(posicion,conversiontriples,color = 'Purple')
plt.savefig("Posicion conversion triples.png")

```
