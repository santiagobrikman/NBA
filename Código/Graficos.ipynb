```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('regular.csv', encoding = 'latin-1')
```

```python
tripleros = df.sort_values(by = '3P', ascending = False).head(5)
tripleros = tripleros.sort_values(by = '3P', ascending = True)
triplerosx = tripleros['Player']
triplerosy = tripleros['3P']
```

```python
fig1 = plt.figure(1)
fig1.set_figheight(10)
fig1.set_figwidth(15)
fig1.set_facecolor('white')
ax = fig1.add_subplot(1,1,1)
plt.title('Jugadores con m√°s cantidad de triples - Temporada regular 2021-2022')
plt.xlabel('Triples por partido')
plt.ylabel('Jugador')
ax.barh(triplerosx, triplerosy, color = '#4CAF50')
plt.savefig("Tripleros.png")
```

```python
edadpromedio = df['Age']
fig2 = plt.figure(1)
fig2.set_figheight(10)
fig2.set_figwidth(15)
fig2.set_facecolor('white')
plt.title('Edad promedio de los jugadores de la NBA - Temporada regular 2021-2022')
plt.xlabel('Jugadores')
plt.ylabel('Edad')
plt.boxplot(edadpromedio)
plt.savefig("Edad Promedio.png")
```

```python
df2 = df.groupby(['Tm']).mean()
df2 = df2.sort_values(by = 'Age', ascending=True).head(5).reset_index()
df2 = df2.sort_values(by = 'Age', ascending=False)

equipo = df2['Tm']
edadpromedio = df2['Age']

fig2 = plt.figure(1)
fig2.set_figheight(10)
fig2.set_figwidth(8)
fig2.set_facecolor('white')
plt.title('Equipos con jugadores de menor edad - Temporada regular 2021-2022')
plt.xlabel('Equipo')
plt.ylabel('Edad promedio')
plt.barh(equipo,edadpromedio,color = '#4CAF50')
plt.savefig("Equipos jovenes.png")

```

```python
df2 = df.groupby(['Tm']).mean()
df2 = df2.sort_values(by = 'Age', ascending=False).head(5).reset_index()
df2 = df2.sort_values(by = 'Age', ascending=True)

equipo = df2['Tm']
edadpromedio = df2['Age']

fig2 = plt.figure(1)
fig2.set_figheight(10)
fig2.set_figwidth(8)
fig2.set_facecolor('white')
plt.title('Equipos con jugadores de mayor edad - Temporada regular 2021-2022')
plt.xlabel('Equipo')
plt.ylabel('Edad promedio')
plt.barh(equipo,edadpromedio,color = '#4CAF50')
plt.savefig("Equipos viejos.png")
```

```python
#Posici√≥n con mejor conversi√≥n de triples

posicion_tripleros1 = df.groupby(['Pos']).mean().reset_index()
posicion_tripleros1 = posicion_tripleros1[['Pos', '3P%']].sort_values(by = '3P%', ascending = False)
posicion_tripleros1 = posicion_tripleros1.sort_values(by='3P%', ascending=True)


posicion = posicion_tripleros1['Pos']
conversiontriples = posicion_tripleros1['3P%']

fig2 = plt.figure(1)
fig2.set_figheight(10)
fig2.set_figwidth(8)
fig2.set_facecolor('white')
plt.title('Posiciones con mayor conversi√≥n de triples - Temporada regular 2021-2022')
plt.xlabel('Conversi√≥n de triples')
plt.ylabel('Posici√≥n')
plt.barh(posicion,conversiontriples,color = 'Purple')
plt.savefig("Posicion conversion triples.png")

```
